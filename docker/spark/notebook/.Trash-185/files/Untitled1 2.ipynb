```python
import json
from kafka import KafkaProducer

# الاتصال بسيرفر كافكا داخل الشبكة المشتركة
producer = KafkaProducer(
    bootstrap_servers=['kafka:9092'],
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# إرسال بيانات الطالب التجريبية
data = {"id": 4, "name": "Mohamed", "age": 23}
producer.send('students', value=data)
producer.flush()

print("✅ تم ضخ بيانات Mohamed داخل توبيك كافكا بنجاح من جوبيتر!")
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("spark://172.22.0.2:7077") \
    .appName("KafkaTest") \
    .config("spark.driver.host", "jupyter") \
    .config("spark.driver.bindAddress", "0.0.0.0") \
    .getOrCreate()
```

```python
spark.range(10).show()
```

```python
df = spark.read \
    .format("kafka") \
    .option("kafka.bootstrap.servers", "kafka:9092") \
    .option("subscribe", "students") \
    .load()

df.printSchema()
```

```python
from pyspark.sql.functions import col

df.select(
    col("key").cast("string"),
    col("value").cast("string")
).show(truncate=False)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("spark://spark-master:7077") \
    .appName("KafkaTest") \
    .config(
        "spark.jars.packages",
        "org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.1"
    ) \
    .getOrCreate()
```

```python
spark.sparkContext.parallelize([1]).count()
```

```python
spark.range(10).show()
```

```python
df = spark.read \
    .format("kafka") \
    .option("kafka.bootstrap.servers", "kafka:9092") \
    .option("subscribe", "students") \
    .option("startingOffsets", "earliest") \
    .load()
```

```python
print(df._jdf.queryExecution().toString())
```
